```python
from matplotlib import pyplot as plt
import cv2
import time
import numpy as np
```

```python
last = time.time()
cat = cv2.imread('cat.2.jpg')
now = time.time()
print(f'image loaded in {now-last} seconds')
last = now
cat = cv2.resize(cat, (32,32), interpolation = cv2.INTER_AREA)
now = time.time()
print(f'image resized in {now-last} seconds')
last = now
plt.imshow(cat)
now = time.time()
print(f'image showed in {now-last} seconds')
```

```python
for name in os.listdir('train'):
    image = cv2.resize(cv2.imread(f'train/{name}'), (32,32), interpolation = cv2.INTER_AREA)
    cv2.imwrite(f'train2/{name}', image)
    print(name)
```

```python
for name in os.listdir('raw_test'):
    image = cv2.resize(cv2.imread(f'raw_test/{name}', cv2.IMREAD_GRAYSCALE), (32,32), interpolation = cv2.INTER_AREA)
    cv2.imwrite(f'test/{name}', image)
    print(name)
```

```python
def refactor(size):
    names = [f'train{size}', f'test{size}']
    if names[0] not in os.listdir('sources'):
        os.mkdir(f'sources/{names[0]}')
        for name in os.listdir('sources/train'):
            image = cv2.resize(cv2.imread(f'sources/train/{name}'), (size,size), interpolation=cv2.INTER_AREA)
            cv2.imwrite(f'sources/{names[0]}/{name}', image)
            print(name)

    if names[1] not in os.listdir('sources'):
        os.mkdir(f'sources/{names[1]}')
        for name in os.listdir('sources/test'):
            image = cv2.resize(cv2.imread(f'sources/test/{name}'), (size,size), interpolation=cv2.INTER_AREA)
            cv2.imwrite(f'sources/{names[1]}/{name}', image)
            print(name)
```
